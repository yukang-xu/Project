I am a big fan of sports and always wondering whether data analysis could navigate team management? I got the answer from one of my favorite movies, Moneyball, in which a small-market team leveraged more analytical gauges of player performance successfully to field a competitive team. My mission in this project is to implement that idea and turn it into an automated recommender system.

Imaging you as a manager of LA Lakers, you just won the championship (Congrats!!!!)  At this moment, you sit at your desk and think about how to continue success. After researching legendary teams in history, you found that stability is their common ground, which means not too much trades and injuries in the middle of season. 

Firstly, let’s take look at the relationship between standing and injuries. Since 2017, the average amount of games missed due to injury in top 5 team is apparently lower than the rest.
